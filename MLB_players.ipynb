```python
import pandas as pd
import numpy as np
```

```python
batters = pd.read_excel("mlb-player-stats-Batters.xlsx")
pitchers = pd.read_excel("mlb-player-stats-P.xlsx")
```

```python
batters[:2]
```

```python
pitchers[:2]
```

```python
batters = pd.DataFrame(batters)
pitchers = pd.DataFrame(pitchers)
```

```python
batters = batters.drop(columns=['SH', 'SF', 'HBP'])
pitchers = pitchers.drop(columns=['CG', 'H', 'SV', 'HLD'])
```

```python
print(batters[:2])
print(pitchers[:2])
```

```python
max_batters = batters[["G","AB","R","H","2B","3B","HR","RBI","SB","BB","SO","AVG","OBP","SLG","OPS"]]
low_batters = batters[["CS"]]
max_pitchers = pitchers[["G","GS","CG","SHO","IP","H","W"]]
low_pitchers = pitchers[["ER","BB","HR","L","BS","ERA","WHIP"]]
```

```python
max_batters.loc[:2]
```

```python
total_max_batters = []
for i in max_batters:
    total_max_batters.append(max_batters[i]/(max(max_batters[i])))
print(total_max_batters)
```

```python
total_max_pitchers = []
for i in max_pitchers:
    total_max_pitchers.append(max_pitchers[i]/(max(max_pitchers[i])))
print(total_max_pitchers)
```

```python
total_low_batters = []
for i in low_batters:
    total_low_batters.append(1-(low_batters[i]/(max(low_batters[i]))))
print(total_low_batters)
```

```python
total_low_pitchers = []
for i in low_pitchers:
    total_low_pitchers.append(1-(low_pitchers[i]/(max(low_pitchers[i]))))
print(total_low_pitchers)
```
